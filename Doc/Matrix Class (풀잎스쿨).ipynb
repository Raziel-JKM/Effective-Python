```python
# C++ 스타일의 행렬 클래스
# 인덱스로 행렬 원소에 접근

class Matrix:
    # matrix_list = [[1, 1], [2, 2]]
    def __init__(self, matrix_list, dtype):
        self.matrix = matrix_list
        self.dtype = dtype
        # 퀴즈 : 이전에 벡터 클래스에서 dtype을 검사했던 것처럼, 
        # 행렬의 데이터 타입 검사를 반복문과 조건문으로 구현해보세요!!
        
        for _, row in enumerate(self.matrix):
            for _, item in enumerate(row):
                if type(item) != self.dtype:
                    print("데이터 타입이 다릅니다.")
                    raise ValueError
    
    def check_dimension_for_mul(self, other):
        # 행렬의 곱셈 : (m x k) X (k x n) -> (m x n)
        
        # 조건문으로 차원을 검사해서 False, True를 반환하는 함수를 만들겠습니다.
        if len(self.matrix[0]) != len(other.matrix):
            print(f"The dimension of two vectors are not same") 
            return False 
        else: 
            return True
        
    def check_dimension(self, other):
        #  두 행렬의 합과 차는 같은 차원의 행렬만 정의가 된다. 
        # (m x n) +- (m x n) -> (m x n)
        
        # 두 행렬의 합과 차를 구할 때 검사를 하는 함수.
        # 반복문과, 조건문으로 차원을 검사해서 False, True를 반환하는 함수를 만들겠습니다.
        if len(self.matrix) == len(other.matrix) and len(self.matrix[0]) == len(other.matrix[0]):
            return True 
        else: 
            print(f"The dimension of two vectors are not same")
            return False
        
    def __add__(self, other):
        # check_dimension()
        matrix = []
        if self.check_dimension(other):
            for row1, row2 in zip(self.matrix, other.matrix):
                row = []
                for col1, col2, in zip(row1, row2):
                    row.append(col1 + col2)
                matrix.append(row)
        else:
            raise ValueError
            
        new_matrix = Matrix(matrix, self.dtype)
        return new_matrix
    
    # 코드 리팩터링
    def __add__2(self, other):
        matrix = []
        if self.check_dimension(other):
            for row1, row2 in zip(self.matrix, other.matrix):
                row = [(col1 + col2) for col1, col2 in zip(row1, row2)]
                matrix.append(row)
        else:
            raise ValueError
            
        new_matrix = Matrix(matrix, self.dtype)
        return new_matrix
        
    def __sub__(self, other):
        matrix = []
        if self.check_dimension(other):
            for row1, row2 in zip(self.matrix, other.matrix):
                row = []
                for col1, col2, in zip(row1, row2):
                    row.append(col1 - col2)
                matrix.append(row)
        else:
            raise ValueError
            
        new_matrix = Matrix(matrix, self.dtype)
        return new_matrix
        
    def __mul__(self, other):
        # check_dimension_for_mul()
        # Hint : 행렬의 곱셈 = (i x k) X (k x j) -> (m x n)
        # 각 반복문에서 어떤 요소를 꺼내올 것인지 고민
        
        matrix = []
        
        for i in range(len(self.matrix)): # 앞의 행렬의 행에 접근한다.
            row = []
            for j in range(len(other.matrix[0])): # 뒤의 행렬의 열에 접근한다. 
                product_sum = 0
                for k in range(len(other.matrix)): # 곱셉 계산을 반복한다. len(other.matrix) or len(self.matrix[0]) 
                    product_sum += self.matrix[i][k] * other.matrix[k][j]
                row.append(product_sum)
            matrix.append(row)    
        new_matrix = Matrix(matrix, self.dtype)
        return new_matrix
    
    def __str__(self):
        msg = ""
        for row in self.matrix:
            msg += f"{row} \n"
        return msg
        
```

```python
mat1 = Matrix([[1, 1], [1, 1]], int)
mat2 = Matrix([[-1, -1], [-1, -1]], int)
mat3 = mat1 + mat2
mat4 = mat1 - mat2
print(mat3)
print()
print(mat4)
print()
```
